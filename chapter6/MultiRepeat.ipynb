```python
for i in range(2,10): # 2~9단까지 => 출력 값이 2단부터이며, 첫 시작이 2 * 2이기에 초기값 수정
    for j in range(2,10): # 각 구구단은 2부터 9까지
        print( i,"*",j,"=",i*j) # 값 출력
    print("********************") # 해당 단이 끝나면 구분선
```

```python
for i in range(1,10): # 총 9줄
    for j in range(i,10): # 각 열에 찍힐 수 -> 거꾸로 출력하기 위해
        print( "*",end="")
    print(end="\n") # 해당 줄에 별을 모두 찍으면 줄바꿈
```

```python
i=1 # 열 카운터 변수
while True: # 무한루프
    j=i # 별 갯수 카운터 변수
    while True: # 무한루프
        if(j==10): break # 거꾸로 출력하기 위해 기준값 도달시 두번째 무한루프 탈출
        print( "*",end="") # 별찍기
        j+=1 # 카운터 변수 증가
    if(i==10): break # 목표 줄 도달시 첫번째 무한루프 탈출
    print(end="\n") # 별을 다 찍었으니 줄바꿈
    i+=1 # 다음줄로 카운터 변수 증가
```

```python
Mon=1
for date in [31,28,31,30,31,30,31,31,30,31,30,31]: # 마지막 날짜들을 넣은 컬랙션으로 for문
    print(Mon,"월 :",end="") # 해당 월 출력
    for Day in range(1,date+1): # 날짜의 나열
        if(Day==date): print(Day, end="\n") # 마지막 날짜에는 , 출력 안하고 줄바꿈 입력
        else: print(Day,end=",") # 줄줄이 날짜 출력
    Mon+=1 # 달 하나 증가
```

```python
for i in range(2,10,3): # 한줄에 3개씩 출력하기 위해 증가는 +3씩, 2단부터 시작
    for j in range(2,10):
        print(i,"*",j,"=",i*j,end="\t") # 첫번째줄 -> 2, 5, 8단
        print(i+1,"*",j,"=",i*j,end="\t") # 두번째줄 -> 3, 6, 9단
        if(i+2!=10): print(i+2,"*",j,"=",i*j,end="\n") # 세번째줄 -> 4, 7단 => 10단이 아닐경우에만 출력
        else: print(end="\n") # 10단 출력 자리일경우? 줄바꿈으로 출력 X
    print(end="\n") # 해당 줄의 단들이 출력 종료후, 한칸띄우기!
```
